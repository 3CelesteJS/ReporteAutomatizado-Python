```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import docx
from docx.shared import Cm
```

```python
df=pd.read_csv('empleados.csv')
```

```python
# docx me permite generar documentos de word
#doc=docx.Document()

# Guardar documento 
doc.save('ReporteAutomatizado.docx')

%config InlineBackend.figure_formats=['svg']# cuando el formato por default que me muestres sean svg(vectores)
# Estetica para las graficas
```

```python
#Descripcion
doc=docx.Document()
doc.add_heading('Graficas')
doc.add_paragraph('Seaborn&Matplotlib')
doc.add_paragraph()

# Como agregar una imagen
p=doc.add_paragraph()
r=p.add_run()
r.add_picture('boxplots_1.jpg',width=Cm(12),height=Cm(9))

# Como agregar una imagen
p=doc.add_paragraph()
r=p.add_run()
r.add_picture('boxplots_salarios.jpg',width=Cm(12),height=Cm(7))

# Como agregar una imagen
p=doc.add_paragraph()
r=p.add_run()
r.add_picture('histograma_equipos.jpg',width=Cm(15),height=Cm(4.5))

# Como agregar una imagen
p=doc.add_paragraph()
r=p.add_run()
r.add_picture('histograma_equipos_genero.jpg',width=Cm(12),height=Cm(7))

# Como agregar una imagen
p=doc.add_paragraph()
r=p.add_run()
r.add_picture('boxplots_subplots.jpg',width=Cm(12),height=Cm(7))

# Como agregar una imagen
p=doc.add_paragraph()
r=p.add_run()
r.add_picture('boxplots_salarios_genero.jpg',width=Cm(12),height=Cm(7))

# Como agregar una imagen
p=doc.add_paragraph()
r=p.add_run()
r.add_picture('relplot.jpg',width=Cm(12),height=Cm(7))

# Como agregar una imagen
p=doc.add_paragraph()
r=p.add_run()
r.add_picture('violin.jpg',width=Cm(12),height=Cm(7))

# Como agregar una imagen
p=doc.add_paragraph()
r=p.add_run()
r.add_picture('violinTrue.jpg',width=Cm(12),height=Cm(7))

# Como agregar una imagen
p=doc.add_paragraph()
r=p.add_run()
r.add_picture('displot.jpg',width=Cm(12),height=Cm(7))


doc.save('prueba.docx')
```

```python
# Funcion figsize para cambiarlo de pulgadas a cm
def cm2pulg(valor):
    return valor/2.54
```

```python
#Grafica

#seaborn usa matplotlib 
## seaborn tiene funciones para construir graficas con matplotlib

fig,axes=plt.subplots(1,2,figsize=(cm2pulg(35),cm2pulg(8))) # Una fila de graficas que tenga 2 columnas

sns.boxplot(ax=axes[0],data=df,x='Equipo',y='Salario') #para agregar figuras 
sns.boxplot(ax=axes[1],data=df,x='Equipo',y='Salario') #para agregar figuras
#Nombre
plt.savefig('boxplots_1.jpg',dpi=300,bbox_inches='tight')
```

```python
#Grafica

fig,axes=plt.subplots(1,2,figsize=(cm2pulg(35),cm2pulg(8))) # Una fila de graficas que tenga 2 columnas

sns.boxplot(ax=axes[0],data=df[df['Definitivo']==True],
            x='Equipo', y='Salario')
sns.boxplot(ax=axes[1],data=df[df['Definitivo']==False],
            x='Equipo', y='Salario')
#Nombre
plt.savefig('boxplots_salarios.jpg',dpi=300,bbox_inches='tight')
```

```python
# Agregar grafica Salario por departamentos mostrando la diferencia
# entre hombres y mujeres
```

```python
# Grafica
sns.set(rc={'figure.figsize':(cm2pulg(12),cm2pulg(6))})
g=sns.countplot(data=df, x='Equipo',palette=sns.color_palette('terrain'))
#  separacion de textos
g.set_xlabel('Departamento',fontsize=15)
g.set_ylabel('N de empleados',fontsize=15)
g.set_xticklabels(g.get_xticklabels(),rotation=70);
#Nombre
plt.savefig('histograma_equipos.jpg',dpi=300,bbox_inches='tight' )
```

```python
# Grafica
sns.set(rc={'figure.figsize':(cm2pulg(12),cm2pulg(6))})
g=sns.countplot(data=df, x='Equipo',hue='Genero',
                palette=['#03a5fc','#c2fc03']) #txt que representa color
#  separacion de textos
g.set_xlabel('Departamento',fontsize=15)
g.set_ylabel('N de empleados',fontsize=15)
g.set_xticklabels(g.get_xticklabels(),rotation=70);
sns.move_legend(g,'upper left',bbox_to_anchor=(1,1))

#Nombre
plt.savefig('histograma_equipos_genero.jpg',dpi=300,bbox_inches='tight' )
```

```python
#Grafica                                                  #1
sns.set(rc={'figure.figsize':(cm2pulg(12),cm2pulg(6))})
g=sns.relplot(data=df, x='Equipo',y='Salario',hue='Genero',
                palette=['#03a5fc','#c2fc03']) #txt que representa color

#Nombre
plt.savefig('relplot.jpg',dpi=300,bbox_inches='tight' )
```

```python
#Grafica                                                  #2
sns.set(rc={'figure.figsize':(cm2pulg(12),cm2pulg(6))})
g=sns.catplot(data=df, x='Equipo',y='Salario',kind='violin',
                palette=['#de1dcb','#ded11d']) 
#Nombre
plt.savefig('violin.jpg',dpi=300,bbox_inches='tight' )
```

```python
#Grafica                                                  #3
sns.set(rc={'figure.figsize':(cm2pulg(30),cm2pulg(30))})
g=sns.catplot(data=df, x='Equipo',y='Salario',kind='violin',hue='Genero',split=True,
                palette=['#34ebe1','#ded11d']) 
#Nombre
plt.savefig('violinTrue.jpg',dpi=300)
```

```python
#Grafica                                                  #4
sns.set(rc={'figure.figsize':(cm2pulg(12),cm2pulg(6))})
g=sns.displot(data=df, x='Salario',kind='kde',hue='Genero',
                palette=['#de1dcb','#ded11d']) #txt que representa color
#Nombre
plt.savefig('displot.jpg',dpi=300,bbox_inches='tight' )
```
